```python
"""
Использование разных моделей машинного обучения для предсказывания диабета.
Использовался датасет Pima Indians Diabetes Database 
https://www.kaggle.com/datasets/uciml/pima-indians-diabetes-database/code?datasetId=228&sortBy=voteCount
Основной метрикой был выбрана recall
Разобраны логистическая регрессия, k ближайших соседей, метод опорных векторов и деревянный бустинг.
Лучше всего себя на момент написания показал метод опорных векторов, но видел результаты бустинга превосходящие его.
"""
```

```python
#Импорты

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import recall_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report
from sklearn.model_selection import GridSearchCV
import lightgbm as lgbm
sns.set()
```

```python
#Читаю файл + предобрабатываю данные(заменяю нули на NaN в столбцах и после меняю NaN на среднее значение столбца)

data = pd.read_csv("D:/ds/pets/diabets/diabetes.csv")

data.info(verbose=True)
print(data.describe())

data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = data[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)

print(data.isnull().sum())
sns.pairplot(data,hue='Outcome')

data['Glucose']=data['Glucose'].fillna(data['Glucose'].mean())
data['BloodPressure']=data['BloodPressure'].fillna(data['BloodPressure'].mean())
data['SkinThickness']=data['SkinThickness'].fillna(data['SkinThickness'].mean())
data['Insulin']=data['Insulin'].fillna(data['Insulin'].mean())
data['BMI']=data['BMI'].fillna(data['BMI'].mean())
print(data.isnull().sum())
```

```python
#Создаю тренировачные и тестовые выборки

datawithouttarget = data.drop(columns=['Outcome'])

scaler = StandardScaler()
datawithouttarget = scaler.fit_transform(datawithouttarget)
datawithouttarget = pd.DataFrame(datawithouttarget, columns=data.columns[:8].tolist())




xtrain, xtest, ytrain, ytest = train_test_split(datawithouttarget, 
                                                data['Outcome'],
                                                test_size=0.25,random_state=42)
```

```python
#Нахожу лучшие гиперпараметры для моделей и делаю кросс валидацию для некоторых из них

param_svc = {'C':np.arange(0.1,10)}
svc = SVC()
svc_grid = GridSearchCV(svc, param_svc, scoring='recall', cv=5)
svc_cv= RandomizedSearchCV(svc,param_svc,cv=5, scoring='recall')
svc_cv.fit(xtrain,ytrain)
svc_grid.fit(xtrain,ytrain)


print("Best Score svc grid:" + str(svc_grid.best_score_))
print("Best Parameters svc grid: " + str(svc_grid.best_params_))
print("Best Score svc:" + str(svc_cv.best_score_))
print("Best Parameters svc: " + str(svc_cv.best_params_))

param_knn = {'n_neighbors':np.arange(1,50)}
knn = KNeighborsClassifier()
knn_cv= RandomizedSearchCV(knn,param_knn,cv=5)
knn_cv.fit(xtrain,ytrain)

print("Best Score knn:" + str(knn_cv.best_score_))
print("Best Parameters knn: " + str(knn_cv.best_params_))



param_lgbm = {'learning_rate' : [0.01, 0.03, 0.05, 0.08, 0.1, 0.2, 0.3, 0.4],
              'n_estimators' : [100, 200, 300, 400, 500],
              'num_leaves': (6, 50),               
              'max_depth': [-1, 1, 2, 3, 4, 5, 6, 7]}

lgbmclf = lgbm.LGBMClassifier(class_weight='balanced',n_jobs=4, verbosity=-1)

lgbmRS = RandomizedSearchCV(lgbmclf, param_lgbm,cv=5,scoring='recall')

lgbmRS.fit(xtrain,ytrain)
parameters = lgbmRS.best_params_
lgbmfinaly = lgbm.LGBMClassifier(**parameters)


knn = knn_cv.best_estimator_
logreg = LogisticRegression(class_weight='balanced')
#svc = SVC(class_weight='balanced', C=svc_cv.best_params_['C'])
svc = SVC(class_weight='balanced', C=1)

scoresknn = cross_val_score(knn, xtrain, ytrain, cv=5, scoring='recall')
scoreslogreg = cross_val_score(logreg, xtrain, ytrain, cv=5, scoring='recall')
scoressvc = cross_val_score(svc, xtrain, ytrain, cv=5, scoring='recall')
```

```python
#Вывод результатов(была взята recall метрика за основу)

print("Оценки на фолдах logreg:", scoreslogreg)
print("Средняя recall logreg:", scoreslogreg.mean())


print("Оценки на фолдах knn:", scoresknn)
print("Средняя recall knn:", scoresknn.mean())

print("Оценки на фолдах svc:", scoressvc)
print("Средняя recall svc:", scoressvc.mean())

knn.fit(xtrain, ytrain)
y_predknn = knn.predict(xtest)
print(f"class report knn:\n{classification_report(ytest,y_predknn)}")


logreg.fit(xtrain, ytrain)
y_pred = logreg.predict(xtest)
print(f"class report logreg:\n{classification_report(ytest,y_pred)}")


svc.fit(xtrain,ytrain)
y_predsvc = svc.predict(xtest)
print(f"class report svc:\n{classification_report(ytest,y_predsvc)}")

lgbmfinaly.fit(xtrain,ytrain)
y_predlgbm = lgbmfinaly.predict(xtest)
print(f"class report lgbm:\n{classification_report(ytest,y_predlgbm)}")
```
